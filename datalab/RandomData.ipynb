## A Random Person

```python
import json as json
import random as random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from generator.person import *
```

```python
with open("data/world-countries.json", 'r') as f1:
    world = json.load(f1)
```

```python
with open("data/nepal-province-district.json", 'r') as f2:
    nepal = json.load(f2)
```

```python
with open("data/profession.json", 'r') as f3:
    prof = json.load(f3)
```

```python
Data = []
for k in range(100):
    P = Person()
    P.set_ID()
    P.set_age()
    P.set_home_address(nepal)
    P.set_foreign_address(world)
    P.set_profession(prof)    

    Data.append({"ID": P.ID,\
                 "age" : P.age,\
                 "home_address" : P.home_address,\
                 "foreign_address": P.foreign_address,\
                 "profession": P.profession})
```
